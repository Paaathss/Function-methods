#LAMDA FUNCTION

lambda functions are similar to user defined functions but without a name .
they are commonly  referred to as anonymous tto a functions .

syntax :

```python
lambda argument (s): expression
```

lamda is a keyword in python for defining the anonymous function .
arguments(s)is a placeholder, that is a variable that will be used to hold the value you want to pass into the function .
a lamda function can have multiple variables depending on what you want to achieve .
expression is the code you want to execute in the lamda function .

```python
#user defined function

def functionnames(n1,n2):
  expression
```

```python
#lamda function
def exp(x):
  return x**2
exp(5)
```

```python
x = lambda x: x.upper() [::-1]
print(x('hello'))
```

```python
x = lambda x: "even" if x%2==0 else "odd"
print(x(32))
```

#FILTER FUNCTION

The filter function operator on a list and returns a subset of the list after applying the filter rule.
The funtion filter(f,,l)needs a function f as it arguments.
f returns a boolean  value ie either true or false .
the function will applied to the every element in the list .
only if f returns True will element of the list l.
only if f returns True will element of the list be included  in the result list.

```python
list1 = [1,2,3,4,5,6,8,9,45,7,99,33]

x = filter(lambda x : x%2==0,list1)
print((x))
```

```python
print(list(x))
```

```python
#another case
print(list(filter(lambda x : x%2==0,list1)))
```

```python
list1 = ['raju','fathi','pathu','nasrin','ke']
x = filter(lambda x : x[0]=='n',list1)
print(list(x))
```

**#map()**

the map() transforms a given list in to new list by transforming each element using a rule,it is used to apply a function to each element of a list or any other iterable .

thi is very useful when we do vector mathematics, the map() function  will take multiple list operators on them return the output list, one thing to notr is that all the list should be of the same size because map() applies the fucntion to corresponding elements  in those lists.


map()

you use the map() function whenever  you want to modify every value in iterable

```python
#syntax of map ()
print(list(map(f,list1)))
```

```python
list1 = [2,4,6]
print(list(map(lambda x : x**2,list1)))
```

```python
list2 = ['raju','ramu','dolu','molu','renu']
print(list(map(lambda name: name.startswith('r'),list2)))
```

#Zip function

function take iterables(can be 0)make iterator that aggregates elements based on theiterables passed,and returns an iterator or tuple.

#syntax
Zip(*iterables)

```python
#[2,3,4],[3,2,3] = ((2,3),(3,2),(4,2))
```

```python
#[2,3,4],[3,2,3],[2,1,4,5] = ((2,3,2),(3,4,1),(4,3,4))
```

```python
#[2,3,4],['raju','bolu','dolu'] = ((2,'raju'),(3,'bolu'),(4,'dolu'))
```

```python
rn = [1,2,3]
name = ['raju','bolu','dolu']
data = list(zip(rn,name))
print(data)
```

```python
rn = [1,2,3]
name = ['raju','bolu','dolu']
data = tuple(zip(rn,name))
print(data)
```

```python
rn = [1,2,3]
name = ['raju','bolu','dolu']
data = dict(zip(rn,name))
print(data)
```
